```python
text = "You say goodbye and I say hello."

text = text.lower()
text = text.replace('.',' .')
print(text)

words = text.split()
print(words)

word_to_id = {}
id_to_word = {} 

for word in words:
  if word not in word_to_id:
    new_id = len(word_to_id)
    word_to_id[word] = new_id  
    id_to_word[new_id] = word

"""딕셔너리에 대한 인지능력이 필요하다. 
강사님의 목적 = 원하는 문장을 라벨링을 하자. 라벨링을 하면 언어를 데이터로 수치화 할 수 있다. 
방법 = 딕셔너리, 반복, 조건문을 사용하자. 새로운 공간을 만들고 포함되지 않았다면 라벨링을 하고, 해당하는 단어가 몇 번째로 들어온 단어인지를 기록시키자.
-> 여기서 내가 정리하고 인지해야 할 것. 1. 딕셔너리에 대한 개념과 저장 방식 2. 반복문으로는 문장에 있는 단어들을 하나씩 꺼내오는데 사용함. 3. 조건문으로는 두 개의 저장공간을 비교해서 저장할 수 있도록 함. 
"""
```
